```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
df = pd.DataFrame(columns=['Title','Company','Experience','Location','URL','Skills','Logo'])
```

```python
url = "https://www.naukri.com/jobs-in-india?experience=0&glbl_qcrc=1028"

driver = webdriver.Chrome("D:\\Selenium\\chromedriver.exe")
driver.get(url)

time.sleep(10)


soup = BeautifulSoup(driver.page_source,'html5lib')
driver.close()
# print(soup.prettify())
results = soup.find(class_='list')
test  = results.find_all(class_='jobTuple')
# test2 = test.find(class_='jobTupleHeader')
# print(test)
# job_elems = results.find_all('article',class_='jobTuple')
# jobTupleHeader">
#    <div class="info fleft">
#     <a class="title ellipsis

for job_elem in test:
    # URL to apply for the job
    jobHeader = job_elem.find(class_='jobTupleHeader')
    jobHeaderInfoFleft = jobHeader.find(class_='info fleft')
    JD = jobHeaderInfoFleft.find(class_='title ellipsis')
    URL = JD.get('href')
    Role = JD.text
    CompName = jobHeaderInfoFleft.find(class_='companyInfo subheading').find(class_='subTitle ellipsis fleft').text
    YOE = jobHeader.find(class_='').find(class_='fleft br2 placeHolderLi experience').find(class_='ellipsis fleft expwdth').text
    LOC = jobHeader.find(class_='').find(class_='fleft br2 placeHolderLi location').find(class_='ellipsis fleft locWdth').text
    Skills = job_elem.find(class_='tags has-description').find_all(class_='fleft dot')
    Logo=None
    # driver.implicitly_wait(10)
    if(jobHeader.find(class_='imgCont fright')!=None):
        print(CompName,jobHeader.find(class_='imgCont fright'))
        if(jobHeader.find(class_='imgCont fright').find(class_='compImg fright')!=None):
            print(CompName,jobHeader.find(class_='imgCont fright').find(class_='compImg fright'))
            Logo = jobHeader.find(class_='imgCont fright').find(class_='compImg fright').get('src')
    skill_list = ""
    for skill in Skills:
        skill_list = skill_list + ","+skill.text
    # URL = job_elem.find('a',class_='jobTuple jobTupleHeader info fleft title ellipsis').get('href')
    # print(Role,CompName,YOE,LOC,skill_list)
    df=df.append({'Title':Role,'Company':CompName,'Experience':YOE,'Location':LOC,'URL':URL,'Skills':skill_list[1:],'Logo':Logo},ignore_index=True)
# print(job_elems)
# print(df)
# print(results.prettify())

```
